```python
df = pd.read_csv('cancer_dataset.csv')
```

```python
x = df.iloc[:,2:]
y = df.iloc[:,1]
```

# HoldOut Validation Approach- Train and Test Split

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
result = model.score(X_test,y_test)
print(result)
```

# K-Fold Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
model = DecisionTreeClassifier()
kfold_validation = KFold(10)
```

```python
results = cross_val_score(model,x,y,cv= kfold_validation)
print(results)
print(np.mean(results))
```

# Stratified K-Fold Cross Validation - Imbalance datasets/classes

```python
from sklearn.model_selection import StratifiedKFold

skfold = StratifiedKFold(n_splits=5)
model = DecisionTreeClassifier()
scores = cross_val_score(model,x,y,cv=skfold )

print(np.mean(scores))
```

```python
print(scores)
```

# Leave One Out Cross Validation(LOOCV)

```python
from sklearn.model_selection import LeaveOneOut
model = DecisionTreeClassifier()
Leave_validation = LeaveOneOut()
results = cross_val_score(model,x,y,cv=Leave_validation)
print(results)
```

```python
print(results.sum())
```

```python
print(np.mean(results))
```

# Repeated Random Train-Test Splits

This technique is a hybrid of traditional train-test splitting  and k-fold cross-validation method. In this technique, we create random splits of the data in the training -test set manner and then repeat the process of splitting and evaluating the algorithm multiple times,just like the cross-validation method.

```python
from sklearn.model_selection import ShuffleSplit
model = DecisionTreeClassifier()
ssplit = ShuffleSplit(n_splits=10,test_size=0.30)
results = cross_val_score(model,x,y,cv=ssplit)
```

```python
results
```

```python
print(np.mean(results))
```
